```python
import pandas as pd
print(pd.__version__)
```

```python
# Reading values from csv
exam_data=pd.read_csv('C:/Users/chait/Desktop/Machine Learning/Datasets/exams.csv',quotechar='"')
exam_data
```

```python
# Calculating average for each column
math_avg=exam_data['math score'].mean()
reading_avg = exam_data['reading score'].mean()
writing_avg= exam_data['writing score'].mean()
print('Math Avg-->',math_avg,'\nReading Avg-->',reading_avg,'\nwriting_avg-->',writing_avg)
```

```python
# Standard normalization for numerical data
from sklearn import preprocessing
exam_data['math score'] = preprocessing.scale(exam_data['math score'])
exam_data['reading score'] = preprocessing.scale(exam_data['reading score'])
exam_data['writing score'] = preprocessing.scale(exam_data['writing score'])
```

```python
# Label Encoding for Categorical data using sklearn
LE=preprocessing.LabelEncoder()
exam_data['gender'] = LE.fit_transform(exam_data['gender'].astype(str))
```

```python
# Label Encoding for Categorical data using pandas
exam_data = pd.get_dummies(exam_data,columns=['gender'])
exam_data 
```
